```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.naukri.com/mnjuser/recommendedjobs")
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup
```

```python
role = "title ellipsis typ-16Bold"
```

```python
company_role_div=soup.find_all(class_=role)
```

```python
comapany_role=[]
for i in company_role_div:
    comapany_role.append(i.text)
```

```python
comapany_role
```

```python
location = "dspIB br2 ver-line placeHolderLi location"
```

```python
location_div=soup.find_all(class_=location)
```

```python
location_names=[]
for i in location_div:
    location_names.append(i.text)
```

```python
location_names
```

```python
company_name = 'dspIB valignM subTitle ellipsis dspIB undefined'
```

```python
name_div=soup.find_all(class_=company_name)
```

```python
comany_names=[]
for i in name_div:
    comany_names.append(i.text)
```

```python
comany_names
```

```python
import csv
```

```python
data = zip(comany_names,comapany_role,location_names)
output_file ="Naukri_Job.csv"
```

```python
with open(output_file,'w',newline='',encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["ComanyName","Designation","Location"])
    writer.writerows(data)
```

```python
print("Data has been saved to",output_file)
```
